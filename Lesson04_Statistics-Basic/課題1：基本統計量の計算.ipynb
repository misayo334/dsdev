```python
import pandas as pd
import numpy as np

#csvファイル読み込み
df = pd.read_csv("./sample_data/california_housing_train.csv")
#中身を表示してみる
df
```

```python
#population列のみ抽出してndarray型にする
population = np.array(df["population"])
#データ型を表示
print(population.dtype)
#中身を表示してみる
population
```

```python
# 合計値
np.sum(population)
```

```python
# 平均値
np.mean(population)
```

```python
# 中央値
np.median(population)
```

```python
# 標本分散(この場合、標本数=populationに含まれる全データ数)
np.var(population, ddof=0)
```

```python
# 不偏分散
np.var(population, ddof=1)
```

```python
# 標本標準偏差
np.std(population, ddof=0)
```

```python
# 不偏標準偏差
np.std(population, ddof=1)
```
